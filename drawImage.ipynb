```python
# encoding=utf-8
%matplotlib inline
import matplotlib.pyplot as plt
from pylab import *         

y1=[]
y2=[]
#read rewards.txt
f1 = open("./trainingResults/rewards_dqn.txt")
f2 = open("./trainingResults/rewards_dueling.txt")
for i in range(0,127):
    line1 = f1.readline()
    line2 = f2.readline()
    y1.append(float(line1.split()[2]))
    y2.append(float(line2.split()[2]))
f1.close()
f2.close()

x = list(range(0,127))
plt.plot(x, y1, mec='r', mfc='w',label='DQN')
plt.plot(x, y2, ms=10, label='Dueling DQN')
plt.legend()  # 
plt.xlabel("Every 100,000 rounds") #X
plt.ylabel("Reward") #Y
plt.title("Comparison of rewards between two algorithms") #

plt.savefig('./trainingResults/results.jpg')
plt.show()
```
